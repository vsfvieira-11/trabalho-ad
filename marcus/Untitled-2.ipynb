```python
import pandas as pd
import numpy as np
import glob
import os

from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score, mean_squared_error

# ====================================================
# PLANILHA
# ====================================================

arquivo_excel = r"C:\Users\marcu\OneDrive\Área de Trabalho\projeto - Análise de Dados.xlsx"

df = pd.read_excel(
    arquivo_excel,
    sheet_name="Amostras selecionadas"
)

base_pls = df[
    ["Código TRABALHO", "Resultado", "CÓD LEC"]
].dropna()

base_pls.columns = [
    "Codigo",
    "Massa_Especifica",
    "Cod_LEC"
]

base_pls["Cod_LEC"] = (
    base_pls["Cod_LEC"]
    .astype(int)
    .astype(str)
)

# ====================================================
# ESPECTROS
# ====================================================

pasta = r"C:\Users\marcu\OneDrive\Área de Trabalho\CSV"

X = []
Y = []
Codigos = []

for _, linha in base_pls.iterrows():

    cod_lec = linha["Cod_LEC"]

    arquivo = os.path.join(
        pasta,
        f"{cod_lec}_tratado.csv"
    )

    if os.path.exists(arquivo):

        espectro = pd.read_csv(arquivo)

        X.append(
            espectro["Absorbancia"].values
        )

        Y.append(
            linha["Massa_Especifica"]
        )

        Codigos.append(
            linha["Codigo"]
        )

X = np.array(X)
Y = np.array(Y)

print("X:", X.shape)
print("Y:", Y.shape)

# ====================================================
# TREINO / TESTE
# ====================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)

# ====================================================
# PLS
# ====================================================

pls = PLSRegression(
    n_components=10
)

pls.fit(
    X_train,
    y_train
)

y_pred = pls.predict(
    X_test
)

# ====================================================
# MÉTRICAS
# ====================================================

r2 = r2_score(
    y_test,
    y_pred
)

rmse = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred
    )
)

print("\nRESULTADOS")
print("R² =", round(r2,4))
print("RMSE =", round(rmse,4))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

plt.scatter(
    y_test,
    y_pred,
    s=60
)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    "r--",
    linewidth=2
)

plt.xlabel("Massa específica real")
plt.ylabel("Massa específica predita")

plt.title(
    f"PLS - Massa Específica\nR² = {r2:.4f}"
)

plt.grid(True)

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score

componentes = range(1, 21)

r2_lista = []

for n in componentes:

    pls = PLSRegression(n_components=n)

    pls.fit(X_train, y_train)

    y_pred = pls.predict(X_test)

    r2 = r2_score(y_test, y_pred)

    r2_lista.append(r2)

melhor_n = componentes[np.argmax(r2_lista)]

print("Melhor número de componentes:", melhor_n)
print("Melhor R²:", max(r2_lista))

plt.figure(figsize=(8,5))

plt.plot(componentes, r2_lista, marker="o")

plt.xlabel("Número de componentes")
plt.ylabel("R²")

plt.title("Otimização do número de componentes PLS")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score, mean_squared_error

# ====================================================
# LEITURA DA PLANILHA
# ====================================================

arquivo_excel = r"C:\Users\marcu\OneDrive\Área de Trabalho\projeto - Análise de Dados.xlsx"

df = pd.read_excel(
    arquivo_excel,
    sheet_name="Amostras selecionadas"
)

# ====================================================
# BASE DE ENXOFRE
# ====================================================

base_enxofre = df[
    ["Código TRABALHO", "Resultado.2", "CÓD LEC.2"]
].dropna()

base_enxofre.columns = [
    "Codigo",
    "Enxofre",
    "Cod_LEC"
]

base_enxofre["Cod_LEC"] = (
    base_enxofre["Cod_LEC"]
    .astype(int)
    .astype(str)
)

print("Total de amostras:", len(base_enxofre))

# ====================================================
# LEITURA DOS ESPECTROS
# ====================================================

pasta = r"C:\Users\marcu\OneDrive\Área de Trabalho\CSV"

X = []
Y = []
Codigos = []

for _, linha in base_enxofre.iterrows():

    arquivo_espectro = os.path.join(
        pasta,
        f"{linha['Cod_LEC']}_tratado.csv"
    )

    if os.path.exists(arquivo_espectro):

        espectro = pd.read_csv(arquivo_espectro)

        X.append(
            espectro["Absorbancia"].values
        )

        Y.append(
            linha["Enxofre"]
        )

        Codigos.append(
            linha["Codigo"]
        )

X = np.array(X)
Y = np.array(Y)

print("Dimensão X:", X.shape)
print("Dimensão Y:", Y.shape)

# ====================================================
# MODELO PLS
# ====================================================

n_componentes = 8

pls = PLSRegression(
    n_components=n_componentes
)

# ====================================================
# VALIDAÇÃO CRUZADA
# ====================================================

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

y_cv = cross_val_predict(
    pls,
    X,
    Y,
    cv=kf
)

# ====================================================
# MÉTRICAS
# ====================================================

r2_cv = r2_score(
    Y,
    y_cv
)

rmse_cv = np.sqrt(
    mean_squared_error(
        Y,
        y_cv
    )
)

print("\nRESULTADOS DO PLS - ENXOFRE")
print(f"R² CV = {r2_cv:.4f}")
print(f"RMSE CV = {rmse_cv:.4f}")

# ====================================================
# GRÁFICO REAL x PREDITO
# ====================================================

plt.figure(figsize=(8,8))

plt.scatter(
    Y,
    y_cv,
    s=60
)

plt.plot(
    [Y.min(), Y.max()],
    [Y.min(), Y.max()],
    "r--",
    linewidth=2
)

plt.xlabel("Enxofre Real")
plt.ylabel("Enxofre Predito")

plt.title(
    f"PLS - Enxofre ({n_componentes} componentes)\nR² = {r2_cv:.4f}"
)

plt.grid(True)

plt.tight_layout()

plt.show()
```
